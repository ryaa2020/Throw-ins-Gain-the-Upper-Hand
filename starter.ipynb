```python
# get location of each throw-in

pass_list = []
for row in new_df.index:
    # print(new_df["location"][row])
    # if (not new_df["pass_type"][row] == "nan"):
    if not pd.isna(new_df["pass_type"][row]):
    # and new_df["pass_type"][row] == "Throw-in":
        pass_list.append([new_df["pass_type"][row], new_df["location"][row]])
        # throwin_location = new_df["location"][row]
print(pass_list)
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
import numpy as np

location_list = []

for item in pass_list:
    if item[0] == 'Throw-in':
        location_list.append(item[1])

p = Pitch(pitch_type='statsbomb')
fig, ax = p.draw(figsize=(12,8))

for loc in location_list:
    p.scatter(loc[0], loc[1], c='orange', s=80, ec='k', ax=ax)
```

```python
import matplotlib.pyplot as plt
# import mplsoccer
from mplsoccer import Pitch, Sbopen
import numpy as np

## load in Statsbomb360 data remotely
parser = Sbopen()
frames, visible = parser.frame(3788741)

## get plotting data
frame_idx = 13
frame_id = visible.iloc[frame_idx].id

visible_area = np.array(visible.iloc[frame_idx].visible_area).reshape(-1, 2)
player_position_data = frames[frames.id == frame_id]

teammate_locs = player_position_data[player_position_data.teammate]
opponent_locs = player_position_data[~player_position_data.teammate]

## set up pitch
p = Pitch(pitch_type='statsbomb')
fig, ax = p.draw(figsize=(12,8))

p.scatter(teammate_locs.x, teammate_locs.y, c='orange', s=80, ec='k', ax=ax)
p.scatter(opponent_locs.x, opponent_locs.y, c='dodgerblue', s=80, ec='k', ax=ax)
p.polygon([visible_area], color=(1, 0, 0, 0.3), ax=ax)

plt.show()
```
