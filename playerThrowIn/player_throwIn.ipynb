# Player Throw-ins

In this notebook, we look at players and their throw-ins. Analyzing every player's throw-in, we deduce their throw-in styles and sucess rates! 

```python
# import libraries
from statsbombpy import sb
import json
import pandas as pd
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
from prettytable import PrettyTable
```

```python
# obtaining the csv file with the event data from all matches
# creating a dataframe with this data
df_allData = pd.read_csv('full_data_sorted.csv')
df_allData
```

```python
df_throwIn = df_allData.loc[df_allData['pass_type'] == 'Throw-in']
df_throwIn
```

```python
df_throwIn_sort = df_throwIn.sort_values(by = "index")
df_throwIn_sort
```

```python
# marking the complete first contact passes
df_throwIn_sort.pass_outcome = df_throwIn_sort.pass_outcome.fillna('Complete')
for row in df_throwIn_sort.index:
    if df_throwIn_sort["pass_outcome"][row]== "Injury Clearance":
        df_throwIn_sort["pass_outcome"][row]="Incomplete"
df_throwIn_sort["pass_outcome"]
```

```python
df_throwIn_clean = df_throwIn_sort.drop(df_throwIn_sort.columns.difference(['pass_outcome', 'location','pass_length','pass_height','pass_angle','player','pass_recipient','team']),1)
df_throwIn_clean = df_throwIn_clean.reset_index()
df_throwIn_clean
```

```python
# list of all teams
teams = []
for team in df_throwIn_clean["team"]:
    if team not in teams:
        teams.append(team)
print(teams)
```

```python
# get one player from each team that throws the most throw ins 
player_info = {}
for row in df_throwIn_clean.index:
    if df_throwIn_clean["player"][row] in player_info:
        player_info[df_throwIn_clean["player"][row]]+=1
    else:
        player_info[df_throwIn_clean["player"][row]]=1
                                                                   
print(player_info)
```

Here, we have a dictionary of every player who has thrown in a ball and the frequency in which they do. 

```python
player_per_team = []
throw_max = 0
throw_player = ""

for team in teams:
    for row in df_throwIn_clean.index:
        if team == df_throwIn_clean["team"][row]:
            current_max = player_info[df_throwIn_clean["player"][row]]
        if current_max > throw_max:
            throw_max = current_max
            throw_player = df_throwIn_clean["player"][row]
    player_per_team.append([throw_player,throw_max,team])
    throw_max=0
print(player_per_team)
```

Now, we have a list of the top throw-in players per team and the frequency.

```python
# Testing the data

# Data for France
i=0
# name of the player
player = player_per_team[i][0]
print(player)

# number of throw ins
throw_count = player_info[player]
print(throw_count)

# number of incomplete throw ins
incomplete_count = 0
for throw in dataframe_collection["France"]["pass_outcome"]:
    if throw == "Incomplete":
        incomplete_count+=1
print(incomplete_count)

# percentage of incomplete throw ins and throw in sucess rate
incomplete_percent = (incomplete_count/throw_count)*100
print(incomplete_percent)
success_rate = 100-incomplete_percent
print(success_rate)

# preferred pass height
low_throw = 0
high_throw = 0
for height in dataframe_collection["France"]["pass_height"]:
    if height == "Low Pass":
        low_throw += 1
    else:
        high_throw+=1
print(low_throw, high_throw)

# preferred pass angle

# angle of throw ins
# the angle of the pass is in radians with 0 poiting towards the teams attacking goal, 
# positive values indivate clockwise angle, negative values indicate counterclockwise values
# lateral : -1<angle<1, backwards < -1, forwards, > 1

lateral_throw = 0
back_throw = 0
forward_throw = 0
for angle in dataframe_collection["France"]["pass_angle"]:
    if angle < -1:
        back_throw +=1
    elif angle>1:
        forward_throw +=1
    else:
        lateral_throw +=1
print(lateral_throw, back_throw, forward_throw)

# preferred pass length
# finding max length of a throw in
max_length = dataframe_collection["France"]["pass_length"].max()
print(max_length)

# finding min length of a throw in 
min_length = dataframe_collection['France']["pass_length"].min()
print(min_length)

# finding the average length of a throw in
avg_length = dataframe_collection['France']["pass_length"].mean()
print(avg_length)

median_length = dataframe_collection['France']["pass_length"].median()
print(median_length)

# define long and short pass : long pass > avg_length , short pass < avg_length
long_throw = 0
short_throw = 0
for length in dataframe_collection['France']["pass_length"]:
    if length >= avg_length:
        long_throw += 1
    else:
        short_throw += 1
print(long_throw, short_throw)
```

```python
player_throwIn_info = {}
i=0
for team in teams:
    # name of the player
    temp_dict = {}
    player = player_per_team[i][0]
    temp_dict["player"] = player

    # number of throw ins
    throw_count = player_info[player]
    temp_dict["throw_count"]=throw_count

    # number of incomplete throw ins
    incomplete_count = 0
    for throw in dataframe_collection[team]["pass_outcome"]:
        if throw == "Incomplete":
            incomplete_count+=1
    temp_dict["incomplete_throw_count"]=incomplete_count

    # percentage of incomplete throw ins and throw in sucess rate
    incomplete_percent = (incomplete_count/throw_count)*100
    # print(incomplete_percent)
    temp_dict["failure_rate"]=round(incomplete_percent,2)
    success_rate = 100-incomplete_percent
    # print(success_rate)
    temp_dict["success_rate"] = round(success_rate,2)

    # preferred pass height
    low_throw = 0
    high_throw = 0
    for height in dataframe_collection[team]["pass_height"]:
        if height == "Low Pass":
            low_throw += 1
        else:
            high_throw+=1
    # print(low_throw, high_throw)
    temp_dict["pass_height"]=[("Low Pass",low_throw), ("High Pass",high_throw)]

    # preferred pass angle

    # angle of throw ins
    # the angle of the pass is in radians with 0 poiting towards the teams attacking goal, 
    # positive values indivate clockwise angle, negative values indicate counterclockwise values
    # lateral : -1<angle<1, backwards < -1, forwards, > 1

    lateral_throw = 0
    back_throw = 0
    forward_throw = 0
    for angle in dataframe_collection[team]["pass_angle"]:
        if angle < -1:
            back_throw +=1
        elif angle>1:
            forward_throw +=1
        else:
            lateral_throw +=1
    # print(lateral_throw, back_throw, forward_throw)
    temp_dict["pass_angle"] = [("lateral", lateral_throw), ("backwards", back_throw), ("forwards", forward_throw)]

    # preferred pass length
    # finding max length of a throw in
    max_length = dataframe_collection[team]["pass_length"].max()
    # print(max_length)

    # finding min length of a throw in 
    min_length = dataframe_collection[team]["pass_length"].min()
    # print(min_length)

    # finding the average length of a throw in
    avg_length = dataframe_collection[team]["pass_length"].mean()
    # print(avg_length)

    median_length = dataframe_collection[team]["pass_length"].median()
    # print(median_length)

    # define long and short pass : long pass > avg_length , short pass < avg_length
    long_throw = 0
    short_throw = 0
    for length in dataframe_collection[team]["pass_length"]:
        if length >= avg_length:
            long_throw += 1
        else:
            short_throw += 1
    # print(long_throw, short_throw)
    temp_dict["pass_length"]=[("Long Pass",long_throw), ("Short Pass", short_throw)]
    player_throwIn_info[team] = temp_dict
    i +=1
    
print(player_throwIn_info)
```

Above shows detailed information about throw-ins per player on a team!
