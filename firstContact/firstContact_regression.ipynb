# First Contact Linear Regression

In this notebook, we aim to do linear regression! Our independent variables are throw-in length, height, angle, and location. Our dependent variable is the throw-in outcome (was the first contact sucessful). We conducted linear regression on each of the individual independent variables and also did a multiple linear regression with all the independent variables. 

```python
# import libraries
from statsbombpy import sb
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm
```

```python
# obtaining the csv file with the event data from all matches
# creating a dataframe with this data
df_allData = pd.read_csv('/Users/rachelyang/Documents/GitHub/StatsBomb/full_data_sorted.csv')
df_allData
```

### Cleaning and Filtering data

```python
df_throwIn = df_allData.loc[df_allData['pass_type'] == 'Throw-in']
df_throwIn
```

```python
# marking the complete first contact passes
df_throwIn.pass_outcome = df_throwIn.pass_outcome.fillna('Complete')
for row in df_throwIn.index:
    if (df_throwIn["pass_outcome"][row]== "Injury Clearance") or (df_throwIn["pass_outcome"][row]== "Unknown"):
        df_throwIn["pass_outcome"][row]="Incomplete"
df_throwIn["pass_outcome"]
```

```python
df_throwIn.drop(df_throwIn.columns.difference(['pass_outcome','location', "pass_height", "pass_angle", "pass_length"]), 1, inplace=True)
df_throwIn_sort = df_throwIn.reset_index()
df_throwIn_sort
```

```python
for row in df_throwIn_sort.index:
    df_throwIn_sort["location"][row] = json.loads(df_throwIn_sort["location"][row])
```

```python
df_split_loc = pd.DataFrame(df_throwIn_sort['location'].tolist(), columns=['x_loc', 'y_loc'])
df_split_loc
```

```python
df_throwIn_sorted = df_throwIn_sort.drop(columns="location")
df_throwIn_final = df_throwIn_sorted.assign(x_loc = df_split_loc["x_loc"], y_loc = df_split_loc["y_loc"])
df_throwIn_final
```

Since we are unable to use categorical data for our linear regression, we use dummy variables for both throw-in height and pass outcome. 

```python
# convert "height" to dummy variable
df_throwIn_dummy = pd.get_dummies(df_throwIn_final, columns=['pass_height'], drop_first=True)
# high pass = 0, low pass = 1
df_throwIn_dummy.rename(columns={'pass_height_Low Pass':'pass_height'},inplace=True)
df_throwIn_dummy
```

```python
#convert "outcome" to dummy variable
df_throwIn_dummy1 = pd.get_dummies(df_throwIn_dummy, columns=['pass_outcome'], drop_first=True)
# complete = 0, incomplete = 1
df_throwIn_dummy1.rename(columns={'pass_outcome_Incomplete':'pass_outcome'},inplace=True)
df_throwIn_dummy1.drop(columns="pass_outcome_Out",inplace=True)
df_throwIn_dummy1
```

```python
# let's see the data
df_throwIn_dummy1
```

### Linear Regression with Throw-in Angle

```python
# linear regression with angle
x_angle = df_throwIn_dummy1[['pass_angle']]
x_angle = sm.add_constant(x_angle)
np.asarray(x_angle)
```

```python
y = df_throwIn_dummy1['pass_outcome']
y
np.asarray(y)
```

```python
model_angle = sm.OLS(y, x_angle)
```

```python
results_angle = model_angle.fit()
```

```python
angle_result = results_angle.summary()
print(angle_result)
```

```python
print(f"coefficient of determination: {results_angle.rsquared}")

print(f"adjusted coefficient of determination: {results_angle.rsquared_adj}")

print(f"regression coefficients: {results_angle.params}")
```

```python
print(f"predicted response:\n{results_angle.fittedvalues}")
```

We expect that for every one radian increase in angle, there is a 0.0014 decrease in the probability of throw-in success. Based off of our r-squared value of 4.8253772105821824e-05, throw-in angle explains an extremely small proportion of the variation in throw-in success. Becuase the p-value is greater than 0.05, our data is not statistically significant. 
This is not the best result! Let's try throw-in length! (hopefully this one is better)
