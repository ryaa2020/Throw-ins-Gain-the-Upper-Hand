```python
# import libraries
from statsbombpy import sb
import json
import pandas as pd
import os
import re
```

```python
# assign directory
directory = 'data/three-sixty'

# list of filenames
file_list = []
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        f_number = re.split("[/.]", f)[2]
        file_list.append(f_number)

file_list
```

```python
# combine all events data into one dataframe

all_events_df = sb.events(match_id = 3788741)

for match in file_list:
    df_1 = sb.events(match_id = int(match))
    df_2 = pd.concat([all_events_df, df_1])
    all_events_df = df_2

all_events_df
```
